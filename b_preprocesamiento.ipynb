```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
r = pd.read_csv("res.csv", sep= ';')
r
```

```python
r.info(verbose=True)
```

```python
r = pd.DataFrame(r)
```

```python
r
```

```python
#Se ha elegido 'EmployeeID' como el índice de la tabla
r = r.set_index('EmployeeID')
```

```python
# Se ordeno la tabla por el indice
r = r.sort_index()
```

Las columnas "EmployeeCount" y "StandardHours" son eliminadas debido a su falta de relevancia en la información.

```python
r.drop(columns=['EmployeeCount'],axis=1,inplace=True)
```

```python
r.drop(columns=['StandardHours'],axis=1,inplace=True)
```

Se han imputado los valores faltantes en la columna 'Attrition' utilizando la función 'fillna'. Estos registros faltantes están relacionados con la condición de que no haya ocurrido renuncias. Además, también se ha aplicado el mismo proceso a las columnas 'retirementType' y 'resignationReason', asegurándonos de que los empleados todavía trabajan en la compañía

```python
r['Attrition'] = r['Attrition'].fillna('No')
```

```python
r['retirementType'] = r['retirementType'].fillna('Worked')
```

```python
r['resignationReason'] = r['resignationReason'].fillna('Worked')
```

```python
#Se ha eliminado la fecha de retiro debido a que no aporta información relevante para el modelo.
r=r.drop('retirementDate', axis=1)
```

```python
modas = r.mode().iloc[0]

# Reemplazar los valores NaN con las modas
r = r.fillna(modas)

# Imprimir el DataFrame con los valores NaN reemplazados por las modas
r
```

```python
r.isnull().sum()
```

```python
r=r.reset_index()
```

```python
r
```

```python
r.to_csv('r2.csv', index=True)
```
