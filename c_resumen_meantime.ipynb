```python
# Importar librerias
import pandas as pd
import b_funciones as fn
```

# <h1>  Preprocesamiento de in_time </h1>

```python
#Carga de datos
en_t=pd.read_csv('BDs\in_time.csv', sep=',')
en_t.head()
```

```python
#Se transformÃ³ todo a formato datetime
en_t=fn.convertir_a_tiempo(en_t)

#Eliminamos Unnamed: 0 fue una variable que se creÃ³ al subir la base
en_t=en_t.drop(['Unnamed: 0'],axis=1)

#Se le hizo la transpuesta
en_t=en_t.T

#Se borraron las filas nulas
en_t=en_t.dropna(how='all')
en_t.head()
```

# <h1>  Preprocesamiento de out_time </h1>

```python
#Carga de datos
sa_t=pd.read_csv('BDs\out_time.csv', sep=',')
sa_t.head()
```

```python
#Se transformÃ³ todo a formato datetime
sa_t=fn.convertir_a_tiempo(sa_t)

#Eliminamos Unnamed: 0 fue una variable que se creo al subir la base
sa_t=sa_t.drop(['Unnamed: 0'],axis=1)

#Se le hizo la transpuesta
sa_t=sa_t.T

#Se borraron las filas nulas
sa_t=sa_t.dropna(how='all')
sa_t
```

<h1> Diferencia y transformaciones

```python
#Se calculÃ³ la resta entre out_time e in_time
diferencias_df = pd.DataFrame(index=en_t.index)

# Iterar sobre las columnas de los DataFrames
for columna in sa_t.columns:
    diferencias_df[columna] = [fn.calcular_diferencia(h1, h2) for h1, h2 in zip(en_t[columna], sa_t[columna])]
```

```python
#Se convierten las unidades del df de diferencia entre hora de salida y llegada de minutos a horas
t=diferencias_df/60

#se convierte a dataframe y se redondea el tiempo, ademÃ¡s de esto sacamos la media del tiempo que trabajÃ³ cada dÃ­a cada empleado
t=pd.DataFrame(round(t.mean(),2))

#Se convierten los datos a flotantes
t=t.astype(float)

#Se nombra la columna mean_time
t=t.rename(columns={0:'mean_time'})

```

```python
# Media de las horas que trabajan los empleados
t.mean()
```

```python
# Se exporta a csv
t.to_csv('Tablas\ t.csv', index=True)
```
