```python
#importacion librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import b_funciones as fn
```

```python
r = pd.read_csv("res.csv", sep= ';')
r.head()
```

```python
# importamos la base creada para el tiempo medio de los trabajadores mediante in_time y out_time
mt = pd.read_csv("t.csv")

# se borra columna por que se crea al subir la informacion
mt=mt.drop(['Unnamed: 0'],axis=1)

mt.head()
```

```python
# Se agrega el id de los trabajadores
mt=mt.reset_index()
mt['index']+=1
mt=mt.rename(columns={'index':'EmployeeID'})
```

```python
#Se agrega el tiempo medio a la base de datos tratada
r=r.merge(mt, on='EmployeeID')
```

```python
#informacion de las columnas (tipo, nulos)
r.info(verbose=True)
```

```python
r['Attrition']
```

```python
r['Attrition']=r['Attrition'].fillna('No')
```

```python
#Se crea lista con variables categoricas
list_cat=list(r.select_dtypes(include=['object']).columns)
```

```python
#Se imputan los datos
r=fn.imputar_f (r,list_cat)
```

```python
#Se vuelve la variable employeeID en entera
r = pd.DataFrame(r)
r['EmployeeID']=r['EmployeeID'].astype('int')
r.head()
```

```python
#Se ha elegido 'EmployeeID' como el índice de la tabla
r = r.set_index('EmployeeID')
r=r.sort_values('EmployeeID')
```

Las columnas "EmployeeCount" y "StandardHours" son eliminadas debido a su falta de relevancia en la información.

```python
r.drop(columns=['EmployeeCount'],axis=1,inplace=True)
r.drop(columns=['StandardHours'],axis=1,inplace=True) # retirementDate
r.drop(columns=['retirementDate'],axis=1,inplace=True)
```

Se han imputado los valores faltantes en la columna 'Attrition' utilizando la función 'fillna'. Estos registros faltantes están relacionados con la condición de que no haya ocurrido renuncias. Además, también se ha aplicado el mismo proceso a las columnas 'retirementType' y 'resignationReason', asegurándonos de que los empleados todavía trabajan en la compañía

```python
#convertimos la base r a un csv
r.to_csv('Tablas\ r2.csv', index=True)
```
