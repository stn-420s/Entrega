```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importar datos
bn = pd.read_csv("Tablas\ r2.csv", sep= ',')
bn.head()
```

```python
#Se separa la variable objetivo del dataframe
y = bn[['Attrition']]
```

```python
#Se separa la variables independientes del dataframe
X = bn.drop('Attrition', axis=1)
X.head()
```

```python
#Se separan las variables categoricas
cat=X[['BusinessTravel','Department','Education','EducationField','Gender','JobLevel', 'JobRole', 'MaritalStatus','JobInvolvement','PerformanceRating', 'EnvironmentSatisfaction', 'JobSatisfaction',
       'WorkLifeBalance','StockOptionLevel','resignationReason','retirementType']]
cat.head()
```

```python
# Se separan las variables numericas
num=bn.drop(cat,axis=1)
num.drop(columns=['Attrition','EmployeeID'],inplace=True)
num.head()
```

```python
# TransformaciÃ³n de la variable respuesta en dos clases ya que es categorica
# 1: Yes
# 0: No

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)

print(y[0:5])

print(le.classes_)
```

```python
#Se convierte y a dataframe
y = pd.DataFrame(y)

# Se nombre la variable
y.rename(columns={0: 'Attrition'}, inplace=True)
```

```python
#Se guardan los datos como csv
num.to_csv('Tablas\ num.csv', index=True)
cat.to_csv('Tablas\cat.csv', index=True)
y.to_csv('Tablas\y.csv', index=True)
```
