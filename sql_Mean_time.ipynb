```python
sa_t=pd.read_csv('out_time.csv', sep=',')
df=sa_t
```

```python
for column in df.columns:
    df[column] = pd.to_datetime(df[column])

for column in df.columns:
    if pd.api.types.is_datetime64_any_dtype(df[column]):
        df[column] = df[column].dt.strftime('%H:%M')
```

```python
df2=df
df2=df2.drop(['Unnamed: 0'],axis=1)
df2
```

```python
df2=df2.T
```

```python
df2=df2.dropna(how='all')
```

```python
df2
```

```python
# Crear una función para calcular la diferencia en minutos entre horas
def calcular_diferencia(hora1, hora2):
    if pd.notnull(hora1) and pd.notnull(hora2):
        try:
            h1, m1 = map(int, hora1.split(':'))
            h2, m2 = map(int, hora2.split(':'))
            return (h2 * 60 + m2) - (h1 * 60 + m1)  # Cambio en el orden
        except ValueError:
            return None
    else:
        return None
```

```python
diferencias_df = pd.DataFrame(index=df1.index)

# Iterar sobre las columnas de los DataFrames
for columna in df1.columns:
    diferencias_df[columna] = [calcular_diferencia(h1, h2) for h1, h2 in zip(df1[columna], df2[columna])]
```

```python
t=diferencias_df/60
```

```python
t=pd.DataFrame(round(t.mean(),2))
```

```python
t=t.astype(float)
```

```python
t=t.rename(columns={0:'mean_time'})
```
